```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from students.dimensionality_reduction import (apply_pca, apply_tsne, apply_umap)
from students.clustering import (run_kmeans, run_hierarchical)
from students.visualization import (plot_representation_comparison, plot_cluster_comparison)
```

```python
data = load_breast_cancer()
X = data.data
y = data.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(X_scaled.shape)
```

```python
pca = apply_pca(X_scaled)
tsne = apply_tsne(X_scaled)
umap = apply_umap(X_scaled)

print(pca.shape, tsne.shape, umap.shape)
```

```python
cluster_labels, centers = run_kmeans(X_scaled, n_clusters=2)
```
